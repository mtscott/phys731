#### In this exercise we will write explicitly the three eigenvalues of the maximum likelihood configuration of a $3\times 3$ GOE matrix. The potential for this ensemble is $V(x) = x^2/2$.

### (a) Let $\lambda_1, \lambda_2,\lambda_3$ be the three maximum likelihood eigenvalues of the $3\times 3$ GOE ensemble in decreasing order. By symmetry we expect $\lambda_3 = -\lambda_1$. What do we expect for $\lambda_2$?

One should expect that $\lambda_2=0$. We present two reasons for this. First of all for the maximum likelihood eigenvalues are the configuration that has maximum probability and the equilibrium position of the eigenvalues. Since $V(x) = x^2/2$, we expect symmetry, and since $\lambda_3 = -\lambda_1$ then $\lambda_2$ should be zero. This would allow $\lambda_2$ to repel the other two equally. The other reason is imagining the eigenvalues are riding the potential. In the equilibrium, we would expect for an odd number of values that one is at the lowest point of the valley.


### (b) Consider Eq. (5.28). Assuming ($\lambda_3 = -\lambda_1$), check that your guess for $\lambda_2$ is indeed a solution. Now write the equation for $\lambda_1$ and solve it.

The generic forumla we want to solve is Eq. (5.28).
\begin{align}
\frac{\partial \mathcal{L}}{\partial \lambda_i} = 0 &\implies V'(\lambda_i) = \frac{2}{N} \sum_{j=i, j\neq i}^{N} \frac{1}{\lambda_i-\lambda_j}
\end{align}
However for our case looking at $\lambda_2$, we have $N=3, V'(\lambda_i) = \lambda_i$, so the equation explicitly is
\begin{align}
V'(\lambda_2) &= \lambda_2\\
&= \frac{2}{3}\left(\frac{1}{\lambda_2 - \lambda_1} + \frac{1}{\lambda_2-\lambda_3}\right)\\

\end{align}

Rewriting we have
\begin{align}
\lambda_2^3 - \lambda_1^2\lambda_2 - \frac{4}{3}\lambda_2 &= 0\\
\lambda_2\left(\lambda_2^2 - \lambda_1^2 - \frac{4}{3}\right) &= 0.
\end{align}

This shows that our guess of $\lambda_2=0$ is one solution of maximum likelihood eigenvalues. Using this and the symmetry condition ($\lambda_1 = -\lambda_3)$, we can write the equation for $\lambda_1$ explicitly,

\begin{align}

Rewriting we have
\begin{align}

\lambda_1 &= \pm 1,
\end{align}
and by symmetry that means the other solution is $\lambda_3 = \mp1$.

### (a) By computing $\text{Re}(\mathfrak{g}(x))$ for $x$ between $0$ and $1$, find $V'(x)$ using Eq. (5.38).
 First recall that $\text{Log}(z) = \text{Re}\left(\log|z|\right) + \text{Im}\left(\log(z)\right) = \log|z| + i\arg(z)$. This means for our argument we have

\begin{align}
 \text{Re}(\mathfrak{g}(x)) &= \text{Re}\log\left(\frac{z}{z-1}\right)\\
 &= \log|z| - \log|z-1|.
\end{align}

 Let $z=x+iy$, then we exploit log properties, the definition of a modulus, and arrive at for $x\in(0,1)$

\begin{align}
\text{Re}(\mathfrak{g}(x)) &= \log|z| - \log|z-1|\\
&= \log|x+iy| - \log|(x-1) +iy|\\
&= \ln \left(\sqrt{x^2 +y^2}\right) - \ln \left(\sqrt{(x-1)^2 +y^2}\right)\\
&= \ln\left|\frac{x}{x-1}\right|,
\end{align}

since $y=0$.

### (b) Compute the Hilbert transform of the uniform density to recover your answer in (a).
From the book, the Hilbert transform of a general density is

\begin{align}
   \mathcal{H}(\rho)(x) &= \text{p.v. } \frac{1}{\pi}\int_{\mathbb{R}} \frac{\rho(\lambda)}{x-\lambda} \text{d}\lambda.
\end{align}
However, since we know that $\rho(\lambda) = \chi_{[0,1]}$, simplifications imply
\begin{align}
    \mathcal{H}(\rho)(x) &= \int_0^1 \frac{1}{x-\lambda} \text{d}\lambda\\
    &= -\frac{1}{\pi} \ln|x-\lambda|\Big|_{\lambda = 1}^0\\
    &= -\frac{1}{\pi} \left(\ln|x-1| - \ln|x-0| \right)\\
    &= \frac{1}{\pi} \ln \left|\frac{x}{x-1}\right|.
\end{align}

One might be worried that answers in (a) and (b) differ by a factor of $\pi^{-1}$, but in the remark on pg. 68, we see that $\text{Re}\left(\mathfrak{g}(x)\right)/\pi$ is also called the Hilbert Transform of $\rho(\lambda)$, which reassures us that we have derived the correct relationship between the two problems.

## 3. Exercise 8.2.1 
#### The family of Student's $t$-distributions, parameterized by the tail exponent $\mu$, is given by the probability density $$P_{\mu}(x) = Z_\mu^{-1}\left(1+\frac{x^2}{\mu}\right)^{-\frac{\mu+1}{2}} \text{ with } Z_\mu^{-1} = \frac{\Gamma\left(\frac{\mu+1}{2}\right)}{\sqrt{\mu\pi}\ \Gamma\left(\frac{\mu}{2}\right)}.$$

### (a) What is the potential $V(x)$ and its derivative $V'(x)$ for these laws?
We use Eq. (8.33) and the definition of $P_\mu(x)$ to find $V(x)$.
\begin{align}
Z_\mu^{-1}\exp\left[-V(x)\right] &= Z_\mu^{-1}\left(1+\frac{x^2}{\mu}\right)^{-\frac{\mu+1}{2}}\\
-V(x) &= \frac{-\mu-1}{2}\log\left(1+\frac{x^2}{\mu}\right)\\
V(x) &= \frac{\mu+1}{2}\log\left(1+\frac{x^2}{\mu}\right)
\end{align}

To find $V'(x)$, we simply take the derivative to get:
\begin{align}
V'(x) &= \frac{(\mu+1)x}{x^2+\mu}
\end{align}

### (b) Using Eq. (8.30), show that for a $t$-distributed variable $x$ we have $$\mathbb{E}\left[\frac{x^2}{x^2+\mu}\right] = \frac{1}{1+\mu}.$$

First, we need to find $f'(x)$ to apply Eq. (8.30).
\begin{align}
\frac{x^2}{x^2+\mu} &= f'(x)V'(x)\\
&= \frac{x}{1+\mu}\frac{(\mu+1)x}{x^2+\mu}
\end{align}

So $f'(x) = \frac{x}{1+\mu} \implies f''(x) = \frac{1}{1+\mu}$. Putting this all together we see
\begin{align}
\mathbb{E}\left[\frac{x^2}{x^2+\mu}\right] &= \mathbb{E}\left[\frac{1}{1+\mu}\right] \\
&=\frac{1}{1+\mu},
\end{align}
since the expectation of a constant is that constant.

### (c) Write the Langevin equation for a Student's $t$-distribution. What is the $\mu\to\infty$ limit of this equation?

The Langevin equation is given by Eq (8.34):
\begin{align}
\text{d}X_t &= \text{d}B_t - \frac{1}{2} V'(X_t)\text{d}t,
\end{align}
where $V'(x) = \frac{(\mu+1)x}{x^2+\mu}$.The only part of the Langevin equation that involves $\mu$ is the potential term, $V'(x)$. We  compare two asymptotics for the potential.

\begin{align}
\lim_{\mu\to\infty}V'(x) &\approx \begin{cases} x & x\ll\sqrt{\mu}\\ \frac{\mu+1}{x} & x\gg\sqrt{\mu}\end{cases}.
\end{align}

Now assuming it is the former case (as $x$ is finite and $\sqrt{\mu}$ is not), we see that this is exactly the potential of the Gaussian or Ornstein - Uhlenbeck SDE. So as $\mu\to\infty$, we have Langevin converges to Ornstein - Uhlenbeck,
\begin{align}
\text{d}X_t &= \text{d}B_t - \frac{1}{2} X_t\text{d}t.
\end{align}

### (d) Simulate your Langevin equation for $\mu = 3,20$ time steps per unit time and run the simulation for $20,000$ units of time. Make a normalized histogram of the sampled values of $X_t$ and compare with the law for $\mu = 3$ given above.

```python
import numpy as np
import numpy.matlib
import matplotlib.pyplot as plt

np.random.seed(23)
```

```python
def potential(mu, x):
    vp = ((mu + 1) * x) / (x **2 + mu)
    return (vp)
```

```python
start = 0.
end = 20000.
step = 0.05
sqstep = np.sqrt(step)
mu = 3

time = np.linspace(start, end, num=400001)
x = np.zeros(time.size + 1)
y = np.zeros(time.size + 1)

noisex = np.random.randn(time.size)
noisey = np.random.randn(time.size)


for i in range(time.size):
    x[i+1] = x[i] + potential(mu, x[i]) * step + noisex[i] * sqstep
    y[i+1] = y[i] + potential(mu, y[i]) * step + noisey[i] * sqstep
    
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12,8))
ax[0].plot(time, x[0:-1], time, y[0:-1])
ax[0].set_title('Langevin Walks')
ax[1].hist(x)
ax[1].hist(y)
ax[1].set_title('Histogram')

plt.show()
```

Here we compared two trials of the Langevin dynamics for $\mu=3$ as well as the histogram of their location. We see the repulsion that is expected between these two particles. Neither of the particles appear to have a Gaussian distribution like Ornstein-Uhlenbeck has in Fig. (8.2). Instead, there is a tendency to escape away from the middle and tend farther away.

### (e) Compared to the Gaussian process (Ornstein - Uhlenbeck), the Student $t$-process has many more short excursions but the long excursions are much longer than the Gaussian ones. Explain this behavior by comparing the function $V'(x)$ in the two cases. Describe their relative small $|x|$ and large $|x|$ behavior.

```python
t = np.linspace(-2.5,2.5,100)
plt.plot(t, potential(3,t))
plt.plot(t, t)
plt.xlabel('Location')
plt.ylabel('Potential')
plt.title('Comparing Potentials for SDEs')
plt.legend(['Langevin', 'Ornstein-Uhlenbeck'])

plt.show()
```

Above, we have plotted the two potentials. We see that there is a linear relationship for the Ornstein-Uhlenbeck potential, but the Langevin is nonlinear. This, in conjuction with the asymptotics performed in part (c), elucidate that for small $|x|$ values, they are rather close to each other. In fact, the Langevin has a more stabilizing potential for small $|x|$, meaning more short excursions. However, when $|x|$ is large, we see that the linear potential has a stronger restoring force. This means that in the Langevin case, although it is unlikely, if a particle were to get out past $\pm1$, then the potential is rather weak so the particle can travel ery far and not feel the pull to the center.
